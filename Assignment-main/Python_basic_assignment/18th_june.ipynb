```python
"""
1. What is the role of the 'else' block in a try-except statement? Provide an example
scenario where it would be useful.

"""
#ANSWER 

"""
The else block in a try-except statement is an optional block that executes 
only if the try block completes successfully, with no exceptions being raised.
It's useful for code that relies on the successful execution of the try block
and doesn't need to handle any exceptions that might occur within it.

"""
#Example of  Validating User Input and Performing an Action:

def calculate_square_root():
  while True:  # Loop until valid input is received
    try:
      number = float(input("Enter a positive number: "))
      if number <= 0:
        raise ValueError("Input must be a positive number")
      result = number**0.5  # Calculate square root
      break  # Exit the loop if successful
    except ValueError as e:
      print(e)  # Print error message for invalid input

  # else block executes only if no exceptions occur in the try block
  else:
    print("The square root of", number, "is:", result)

calculate_square_root()

```

```python
"""
2. Can a try-except block be nested inside another try-except block? Explain with an
example.

"""
#ANSWER 
"""
Yes, absolutely! You can nest try-except blocks inside one another in Python.

"""
#EXAMPLE 
def read_and_process_file(filename):
    try:
        # Attempt to open the file for reading
        with open(filename, 'r') as file:
            try:
                # Read data from the file
                data = file.read()
                # Process the data (e.g., convert to uppercase)
                processed_data = data.upper()
                print(f"Processed data: {processed_data}")
            except UnicodeDecodeError:
                print("Error: Unable to decode file content")
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found")

# Example usage:
read_and_process_file("mydata.txt")  # Assuming "mydata.txt" exists
read_and_process_file("nonexistent.txt")  # Assuming "nonexistent.txt" does not exist


```

```python
"""
7. What are log formatters in Python logging, and how can you customise the 
log message format using formatters?

"""
#ANSWER 

"""

In Python logging, log formatters determine the layout and structure of your 
log messages. They specify how the various parts of a log record, such as 
timestamps, log levels, names of loggers, and the actual message, are 
combined into a single string that gets written to the output destination 
(e.g., console, file).

#Customizing Log Messages with Formatters:

The logging module provides the Formatter class, which you can use to define 
the format of your log messages. By default, the basicConfig() function sets 
a simple formatter that includes the timestamp, log level, logger name, and 
the message itself. However, you can create custom formatters to include 
additional information or structure your logs in a way that best suits your 
needs.

#Using String Formatting:

The Formatter class allows you to use Python's built-in string formatting syntax (% formatting) to specify the elements you want to include in the log message and how they should be arranged. Here are some common formatting options:

%(name)s: Name of the logger that generated the message.

%(levelname)s: Text representation of the log level (e.g., DEBUG, INFO, 
WARNING).

%(asctime)s: Timestamp of the log message. You can customize the timestamp 
format using strftime directives (e.g., %(asctime)s:%(msecs)d).

%(message)s: The actual message text that you logged.

"""
#EXAMPLE

import logging

custom_format = "%(asctime)s - %(levelname)s - %(name)s - %(process)d - %(message)s"
formatter = logging.Formatter(custom_format)

# Configure a logger to use the custom formatter
logger = logging.getLogger(__name__)
logger.setLevel(logging.DEBUG)
handler = logging.StreamHandler()
handler.setFormatter(formatter)
logger.addHandler(handler)

logger.debug("This is a debug message with additional details.")
```
