```python
#5. How many global scopes are there in a Python program? How many local scopes?

"""
In a Python program:

There's one global scope per program execution.

There can be potentially many local scopes.

1.Global Scope:

The global scope encompasses the entire Python program. 

It's the outermost scope where variables and functions defined outside of 
any function become accessible.

All code outside of function definitions belongs to the global scope.

The global scope exists as long as the program is running. Once the program 
terminates, the global scope and its variables are forgotten.

2.Local Scope:

A local scope exists within each function definition. It's a more limited 
scope where variables and functions defined within the function are
accessible.

Local scopes are created when a function is called and destroyed when the 
function call finishes execution.

Variables defined within a function are local to that function and cannot be 
accessed directly from outside the function (unless explicitly made global).

"""
#Example 
x = 10  # Global variable

def my_function():
  y = 20  # Local variable

  print(f"Inside function: x = {x}, y = {y}")  # Accessing both global and local variables

# Global scope
print(f"Outside function: x = {x}")  # Can access global variable
# print(f"Outside function: y = {y}")  # This would cause an error (y is local)

my_function()
```

```python
#6. What happens to variables in a local scope when the function call returns?

"""
When a function call in Python returns, the variables in the function's 
local scope are destroyed and become inaccessible. This is because the 
function's local scope is temporary and exists only for the duration of the 
function call.

Here's a breakdown of what happens:

1.Function Call: When you call a function, Python creates a new local scope 
for that function execution. Any variables defined within the function's body
belong to this local scope.

2.Memory Allocation: Memory is allocated for the function's local variables to 
store their values.

3.Function Execution: The function's code block runs, using and potentially 
modifying the values in its local variables.

4.Function Return: When the function reaches a return statement or finishes 
executing its code block, the function returns.

5.Local Scope Destruction: The local scope created for the function call is 
destroyed along with the memory allocated for its variables.

"""
#Example 
def get_square(num):
  """Calculates and returns the square of a number."""
  square = num * num  # Local variable
  return square

result = get_square(5)  # Function call
print(result)  # Output: 25 (square is returned and accessible)

# This would cause an error because "square" is local to the function
# print(f"Outside function: square = {square}")

```

```python
"""
Question 7. What is the concept of a return value? Is it possible to have a 
return value in an expression?

"""
#ANSWER

"""
Understanding Return Values:

1.Function Definition: When you define a function, you can optionally include
a return statement within its code block. This statement specifies the value 
(or data) the function will send back when it finishes running.

2.Function Call: When you call a function, Python executes the function's 
code block. If a return statement is present, it evaluates the expression 
following return and stores the resulting value.

3.Passing the Value Back: After execution, the function "returns" the stored 
value back to the line in your code where you called the function.You can 
typically assign this returned value to a variable to capture it for further 
use.

"""
#Example 
def add(x, y):
  """Adds two numbers and returns the sum."""
  sum = x + y
  return sum

result = add(5, 3)  # Function call, stores returned value in "result"

print(result)  # Output: 8 (returned sum from the function)

```

```python
#9. How do you make a function variable refer to the global variable?

"""
1. Using the global Keyword:

The global keyword explicitly declares that you want to modify a variable
defined in the global scope from within the function.

It's generally recommended to avoid using global excessively, as it can make 
code less readable and harder to maintain. However, there are situations 
where it's necessary.

"""
#Example
count = 0  # Global variable

def increment():
  global count  # Declare using global keyword
  count += 1

increment()
increment()

print(count)  # Output: 2 (global count is modified)

"""
2. Returning the Modified Value (Preferred Approach):

A more preferred approach is to modify the global variable indirectly by 
returning the modified value from the function. This keeps the function's 
logic self-contained and avoids relying on global variables.

You can then assign the returned value to the global variable in the calling 
code.

"""
#Example 
count = 0  # Global variable

def increment():
  count += 1  # Modify local copy (doesn't affect global directly)
  return count

count = increment()  # Call function and assign returned value to global count
count = increment()  # Call again and update global count

print(count)  # Output: 2 (global count is modified indirectly)
```

```python
"""
Question 13. What can you do to save a programme from crashing if it 
encounters an error?

"""
#Answer

"""
There are several techniques you can employ to prevent your program from 
crashing when it encounters errors in Python:

1. Try-Except Blocks:

This is a fundamental approach to error handling in Python. It allows you to
define a code block (the try clause) that might raise an exception (error) 
and one or more except clauses to handle different types of exceptions 
gracefully.

"""
#Example 

def divide(numerator, denominator):
  """Divides two numbers and handles potential ZeroDivisionError."""
  try:
    result = numerator / denominator
  except ZeroDivisionError:
    result = "Error: Cannot divide by zero"
  return result

# Example usage (safe division)
number = 10
divisor = 2
safe_result = divide(number, divisor)
print(f"Safe division: {number} / {divisor} = {safe_result}")

# Example usage (attempting division by zero)
divisor = 0
unsafe_result = divide(number, divisor)
print(f"Unsafe division: {number} / {divisor} = {unsafe_result}")

"""
2. Logging:

While not directly preventing crashes, logging errors can be crucial for 
debugging and monitoring the health of your program. You can use libraries 
like logging to record error messages and stack traces for later analysis.

"""
#Example
import logging

# Configure logging (optional, adjust log level and destination as needed)
logging.basicConfig(level=logging.DEBUG, filename='program_errors.log')

def divide(numerator, denominator):
  """Divides two numbers and handles potential ZeroDivisionError."""
  try:
    result = numerator / denominator
  except ZeroDivisionError as e:
    logging.error("Attempted division by zero", exc_info=True)  # Log error with traceback
    result = "Error: Cannot divide by zero"
  return result


number = 10
divisor = 2
safe_result = divide(number, divisor)
print(f"Safe division: {number} / {divisor} = {safe_result}")

divisor = 0
unsafe_result = divide(number, divisor)
print(f"Unsafe division: {number} / {divisor} = {unsafe_result}")


"""
3. Assertions:

Assertions are a way to test the validity of assumptions in your code. 
You can use the assert keyword followed by a condition. If the condition is 
False, an AssertionError is raised.

"""
#Example 
def validate_age(age):
  assert age >= 18, "Age must be 18 or older."
  # Rest of the function logic

validate_age(25)  # No error
validate_age(12)  # Raises AssertionError with the message
```

```python
"""
14. What is the purpose of the try clause? What is the purpose of the except
clause?

"""
#Answer

"""
1.Try Clause:

The try clause is where you place the code that might potentially raise an 
exception (error) during execution.

This code block can contain any valid Python statements, function calls, or 
operations.

The goal is to execute this code under normal conditions, but if an error 
occurs within the try block, the control flow jumps to the except clause(s) 
to handle the error.

2.Except Clause:

The except clause(s) are used to handle exceptions that are raised within the
try block.

You can have one or more except clauses, each potentially targeting specific 
exception types.

When an exception occurs inside the try block, Python checks the except 
clauses one by one (from top to bottom) to see if the exception type matches 
the exception type specified in the except clause.

If a match is found, the code block within that particular except clause 
executes, and the program continues after the try-except block.

If no matching except clause is found for the raised exception, the exception
propagates upwards, potentially causing the program to crash.

"""
#Example 

def divide(numerator, denominator):
  """Divides two numbers and handles potential ZeroDivisionError."""
  try:
    result = numerator / denominator
  except ZeroDivisionError:
    result = "Error: Cannot divide by zero"
  return result

# Example usage
number = 10
divisor = 2
safe_result = divide(number, divisor)
print(f"Safe division: {number} / {divisor} = {safe_result}")

divisor = 0
unsafe_result = divide(number, divisor)
print(f"Unsafe division: {number} / {divisor} = {unsafe_result}")

```
