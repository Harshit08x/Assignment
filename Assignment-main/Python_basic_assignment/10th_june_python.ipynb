```python
"""
2. How can you pass arguments to a function in Python? Explain the difference 
between positional arguments and keyword arguments.

"""
#ANSWER 

"""
In Python, you can pass arguments to functions to provide them with data to 
work on. There are two main ways to pass arguments: positional arguments and 
keyword arguments.

1.Positional arguments

Arguments are passed in the order they are defined in the function's parameter 
list.

When calling the function, you provide values in the same order.

If you miss an argument or provide an incorrect number of arguments, you'll 
get an error.

"""
#EXAMPLE 
def calculate_area(length, width):
  """This function calculates the area of a rectangle."""
  return length * width

# Pass arguments in the order they are defined in the function
area_1 = calculate_area(5, 3)  # length, then width
print(f"Area using positional arguments: {area_1}")  # Output: Area using positional arguments: 15

"""
2.Keyword arguments

Arguments are passed using keyword-value pairs where the keyword is the 
argument name from the function definition.

This makes the order irrelevant as long as you provide all the required 
keywords.

"""
#EXAMPLE 

# Pass arguments by keyword (argument name = value)
area_2 = calculate_area(width=4, length=6)  # Notice the order doesn't matter here
print(f"Area using keyword arguments: {area_2}")  # Output: Area using keyword arguments: 24
```

```python
"""
3. What is the purpose of the return statement in a function? Can a function have multiple return
statements? Explain with an example.

"""
#ANSWER 

"""
The return statement in Python serves two main purposes in functions:

1.Exiting the function: It signals the end of the function's execution and 
returns a value (or None if no value is explicitly specified) back to the 
calling code. The function stops executing any code after the return statement 
is encountered.

2.Providing a result: The return statement allows you to send back a value 
(or object) from the function as its output. This value can then be used in 
the code that called the function.

Can a function have multiple return statements?

Yes, a function can have multiple return statements throughout its body. 
The function will exit as soon as it encounters a return statement, and only 
the value specified in that particular return statement will be returned.

"""
#EXAMPLE 
def calculate_area(length, width):
  """This function calculates the area of a rectangle and provides informative messages."""

  # Check if both arguments are positive
  if length > 0 and width > 0:
    area = length * width
    return area  # Early return if positive values
  else:
    return "Invalid input: Please provide positive values for length and width."

# Example usage (positive values)
area_1 = calculate_area(5, 3)
print(f"Area (positive values): {area_1}")  # Output: Area (positive values): 15

# Example usage (non-positive values)
area_2 = calculate_area(-2, 4)
print(f"Area (non-positive values): {area_2}")  # Output: Area (non-positive values): Invalid input: Please provide positive values for length and width.
```

```python
"""
4. What are lambda functions in Python? How are they different from regular functions? Provide an
example where a lambda function can be useful.

"""
#ANSWER 

"""
Lambda functions in Python are a concise way to define anonymous functions, 
meaning they are defined without a name. They are typically used for short, 
throwaway functions or when working with higher-order functions (functions 
that take other functions as arguments).

Here's a breakdown of the key differences between lambda functions and regular 
functions:

Regular functions:

 Defined using the def keyword followed by the function name and parentheses for 
 arguments.

 Can have multi-line docstrings and code blocks.

 More readable for complex logic.

Lambda functions:

 Defined using the lambda keyword followed by arguments and a colon (:), then 
 the expression to be evaluated.

 Limited to a single expression.

 Concise for simple operations.

"""
#EXAMPLE OF LAMBDA FUNCTION 
square_lambda = lambda x: x * x

result_lambda = square_lambda(5)
print(f"Result using lambda function: {result_lambda}")  # Output: Result using lambda function: 25

#EXAMPLE OF REGULAR FUNCTION 
def square(x):
  """This function squares a number."""
  return x * x

result_reg_func = square(5)
print(f"Result using regular function: {result_reg_func}")  # Output: Result using regular function: 25

```

```python
"""
6. How can you use the "return" statement in a Python function to return 
multiple values?

"""
#ANSWER 

"""
In Python, there's no built-in way to return multiple values using a single 
return statement. However, you can achieve this using a couple of alternative 
approaches:

Returning a tuple:

 A tuple is an ordered collection of elements enclosed in parentheses ().

 You can pack the desired values you want to return from the function into a 
 tuple using a comma-separated list within the return statement.

 When the function returns, the calling code receives the tuple containing the 
 values.

"""
#EXAMPLE 
def calculate_area_and_perimeter(length, width):
  """This function calculates and returns the area and perimeter of a rectangle."""

  area = length * width
  perimeter = 2 * (length + width)
  return area, perimeter  # Returning a tuple with both values

# Unpacking the returned tuple into separate variables
rectangle_area, rectangle_perimeter = calculate_area_and_perimeter(5, 3)

print(f"Area: {rectangle_area}, Perimeter: {rectangle_perimeter}")

```

```python
"""
7. What is the difference between the "pass by value" and "pass by reference" 
concepts when it comes to function arguments in Python?

"""
#ANSWER 

"""
Pass by value:

 In traditional pass by value, a copy of the argument value is passed to the 
 function.

 Any modifications made within the function only affect the copy, not the 
 original variable outside the function.

Pass by reference:

 In pass by reference, a reference or pointer to the original argument is 
 passed to the function.
 
 Changes made within the function directly affect the original variable.

"""
#EXAMPLE 
# Immutable object (string) - pass by value behavior
def change_greeting(message):
  message = "Hello!"  # Assigning a new string value to the local variable

greeting = "Hi"
change_greeting(greeting)
print(f"Greeting after function call (immutable): {greeting}")  # Output: Greeting after function call (immutable): Hi

# Mutable object (list) - pass by reference behavior
def modify_list(data_list):
  data_list.append(100)  # Modifying the original list object

my_list = [1, 2, 3]
modify_list(my_list)
print(f"List after function call (mutable): {my_list}")  # Output: List after function call (mutable): [1, 2, 3, 100]
```

```python
"""
8. Create a function that can intake integer or decimal value and do following 
operations:
a. Logarithmic function (log x)
b. Exponential function (exp(x))
c. Power function with base 2 (2^x)
d. Square root

"""
#ANSWER 

import math

def calculate(x):
  """
  This function performs various mathematical operations on a numerical input (integer or decimal).

  Args:
      x: The numerical value (integer or decimal).

  Returns:
      A dictionary containing the results of the following operations:
          - log(x): Logarithm of x (base 10).
          - exp(x): Exponential of x (base e).
          - pow(2, x): x raised to the power of 2.
          - sqrt(x): Square root of x.
  """

  results = {
    "log(x)": math.log10(x) if x > 0 else None,  # Handle non-positive values for log
    "exp(x)": math.exp(x),
    "2^x": math.pow(2, x),
    "sqrt(x)": math.sqrt(x) if x >= 0 else None,  # Handle negative values for sqrt
  }
  return results


value = 4  # You can change this value to any integer or decimal

# Get the results dictionary
operations_results = calculate(value)


# Print the results individually with clear labels
for operation, result in operations_results.items():
  if result is not None:
    print(f"{operation}: {result:.4f}")  # Format output to 4 decimal places
  else:
    print(f"{operation}: Not applicable (input value does not meet requirements)")
```

```python
"""
9. Create a function that takes a full name as an argument and returns first 
name and last name.

"""
#ANSWER 

def split_name(full_name):
  """
  This function splits a full name into first name and last name.

  Args:
      full_name: The full name string.

  Returns:
      A dictionary containing the first name and last name, or None if the name cannot be split.
  """

  # Split the name at the first whitespace (space, tab, etc.)
  name_parts = full_name.strip().split(None, 1)

  if len(name_parts) >= 2:
    return {
      "first_name": name_parts[0],
      "last_name": name_parts[1]
    }
  else:
    return None

# Example usage (change the name as needed)
full_name = "Alice Bobcat"
name_parts = split_name(full_name)

if name_parts:
  print(f"First name: {name_parts['first_name']}")
  print(f"Last name: {name_parts['last_name']}")
else:
  print("Could not split the name into first and last name.")
```
