```python
"""
1. What is the role of try and exception block?

"""
#ANSWER 

"""
1. The try Block:

 This block contains the code you want to execute.

 If the code runs without errors, the except block is skipped, and the program 
 continues as usual.

 2. The except Block:

 This block defines how to handle exceptions that may occur within the try 
 block.

 An exception is an event that disrupts the normal flow of your program, such 
 as a division by zero error, trying to access a non-existent list element, 
 or encountering a file that doesn't exist.

 When an exception occurs within the try block, control jumps to the except 
 block.


"""
```

```python
"""
2. What is the syntax for a basic try-except block?

"""
#ANSWER 

try:
  # Your code that might raise exceptions goes here
except:
  # Code to handle any exception (not recommended generally)
  print("An error occurred!")
except SpecificException:
  # Code to handle specific exceptions
  print("Handling specific exception:", SpecificException)
else:
  # Code that executes only if no exceptions occur in the try block
  print("The code executed successfully!")
finally:
  # Code that always executes, regardless of exceptions
  print("This code always executes!")


  
```

```python
"""
3. What happens if an exception occurs inside a try block and there is no 
matching except block?

"""
#ANSWER 

"""
There are two main possibilities depending on the structure of your code:

1.Enclosing try-except Block:

 If there's an enclosing try-except block with a matching except clause, that 
 except block will take over the responsibility of handling the exception. 
 This can be:
    A specific except block that matches the exact exception type (e.g., except 
    ZeroDivisionError).

    A bare except: block, which can handle any type of exception (generally not 
    recommended as it can mask specific errors).

 Once the exception is handled by the matching except block, the program 
 execution continues from there, potentially resuming normal operation or 
 performing specific actions based on the exception type.

2.No Enclosing try-except Block:

If there's no enclosing try-except block with a matching except clause, the 
program encounters an unhandled exception. This leads to a program crash with 
an error message. The message typically includes details like:

  The type of exception that occurred (e.g., ZeroDivisionError)

  The line number where the exception was raised

  A traceback, which shows the call stack leading to the error, helping you 
  pinpoint the exact location where the exception originated.


"""
```

```python
"""
4. What is the difference between using a bare except block and specifying a 
specific exception type?

"""
#ANSWER 

"""
1.Bare except block (except:):

Catches any exception that occurs within the try block.

Pros:

 Simple and concise syntax.

 Ensures the program doesn't crash due to unhandled exceptions.

Cons:

 Lacks specificity: You lose information about the exact type of exception that 
 occurred. This can make debugging more difficult as you can't determine the 
 precise cause of the error.

 Potential masking: It can potentially mask specific exceptions you might want 
 to handle differently. For example, if you have an except block that catches 
 Exception (the base class for most exceptions), it would also catch 
 KeyboardInterrupt exceptions, which might interrupt the program even if it's
 working correctly.

2.Specifying a specific exception type (except SpecificException):

Catches only exceptions of the specified type or its subclasses. For example, 
except ZeroDivisionError will only handle ZeroDivisionError exceptions.

Pros:

 Clarity and control: You have more control over what exceptions get handled 
 and can provide more specific error messages to the user.

 Improved debugging: By knowing the exact exception type, you can more easily 
 pinpoint the source of the error.

Cons:

 Requires you to anticipate potential exception types.
 
 Might require multiple except blocks to handle different exceptions.


"""
```

```python
"""
5. Can you have nested try-except blocks in Python? If yes, then give an example.

"""
#ANSWER 
"""
Yes, Python absolutely allows nesting try-except blocks. This can be very useful for handling 
exceptions at different levels in your code and providing more granular error handling.

"""
#EXAMPLE 

def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Error: Cannot divide by zero")
    else:
        try:
            # Attempt to convert the result to an integer
            int_result = int(result)
        except ValueError:
            print(f"Error: Cannot convert {result} to an integer")
        else:
            print(f"Result: {int_result}")

divide(10, 2)  # Result: 5
divide(10, 0)  # Error: Cannot divide by zero
divide(10, "2")  # Error: Cannot convert 5.0 to an integer
```

```python
"""
6. Can we use multiple exception blocks, if yes then give an example.

"""
#ANSWER 
"""
Yes, you can definitely use multiple except blocks within a single try block in Python. 
This allows you to handle different types of exceptions in a specific way 

"""
#EXAMPLE 

def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Error: Cannot divide by zero")
    except TypeError:
        print("Error: Invalid data type for division")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    else:
        print(f"Result: {result}")


divide(10, 2)  # Result: 5.0
divide(10, 0)  # Error: Cannot divide by zero
divide(10, "2")  # Error: Invalid data type for division
divide(10, [])  # An unexpected error occurred: unsupported operand type(s) for /: 'int' and 'list'
```

```python
"""
8. Write code for the following given scenario and add try-exception block to it.
a. Program to divide two numbers
b. Program to convert a string to an integer
c. Program to access an element in a list
d. Program to handle a specific exception
e. Program to handle any exception

"""
#ANSWER 

#a. Program to Divide Two Numbers:
def divide_numbers(a, b):
    try:
        result = a / b
        print(f"Result of {a} divided by {b} is: {result}")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero")

divide_numbers(10, 2)  # Result: 5.0
divide_numbers(10, 0)  # Error: Cannot divide by zero


#b. Program to Convert a String to an Integer:
def convert_to_integer(s):
    try:
        num = int(s)
        print(f"Converted integer value: {num}")
    except ValueError:
        print(f"Error: Cannot convert '{s}' to an integer")


convert_to_integer("123")  # Converted integer value: 123
convert_to_integer("abc")  # Error: Cannot convert 'abc' to an integer


#c. Program to Access an Element in a List:
def access_list_element(my_list, index):
    try:
        value = my_list[index]
        print(f"Value at index {index}: {value}")
    except IndexError:
        print(f"Error: Index {index} is out of range")


my_list = [1, 2, 3]
access_list_element(my_list, 1)  # Value at index 1: 2
access_list_element(my_list, 10)  # Error: Index 10 is out of range


#d. Program to Handle a Specific Exception:
def specific_exception_handling():
    try:
        # Some operation that may raise a specific exception
        raise FileNotFoundError("File not found")
    except FileNotFoundError as e:
        print(f"Specific exception caught: {e}")

specific_exception_handling()  # Specific exception caught: [Errno 2] File not found


#e. Program to Handle Any Exception:
def handle_any_exception():
    try:
        # Some operation that may raise any exception
        raise Exception("An unexpected error occurred")
    except Exception as e:
        print(f"Exception caught: {e}")


handle_any_exception()  # Exception caught: An unexpected error occurred


```
