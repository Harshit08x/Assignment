```python
"""
Question 2. Discuss the concept of single inheritance and multiple inheritance, highlighting their 
differences and advantages.

Answer - 1.Single Inheritance:

*Definition:

Single inheritance refers to the scenario where a subclass (or derived class) inherits 
from only one base class (or parent class).

*Advantages:

Simplicity:
Single inheritance simplifies the class hierarchy, making it easier to understand and 
maintain.

Clear Relationships: 
Each subclass has a clear parent, and the relationship is straightforward.

*Disadvantages:

Tight Coupling:
A major disadvantage is tight coupling between the parent and child class.
Changes in the parent class can ripple through to child classes, even if unintended. 
This can make maintaining and debugging code more complex.

Limited Flexibility:  
Since a child class inherits everything from the parent, it can be inflexible. 
The child class might inherit functionalities it doesn't need, or changes made to the 
parent might not be reflected in the desired way in the child 

"""
#Example of single inheritance 

class Animal:
  def __init__(self, name):
    self.name = name

  def make_sound(self):
    print("Generic animal sound")

class Dog(Animal):
  def __init__(self, name, breed):
    super().__init__(name)  # Call the parent class constructor
    self.breed = breed

  def make_sound(self):
    print("Woof!")

# Create an Animal object (not very useful on its own)
animal = Animal("Generic Animal")
animal.make_sound()  # Output: Generic animal sound

# Create a Dog object, inheriting from Animal
dog = Dog("ALEX", "Labrador")
dog.make_sound()  # Output: Woof!

# Dog also inherits the name property from Animal
print(dog.name)  # Output: ALEX

"""
2.Multiple Inheritance:

Definition: Multiple inheritance occurs when a subclass inherits from two or more base 
classes.

*Advantages:

Code Reusability: Multiple inheritance allows combining features from different sources 
(base classes).

Richer Functionality: Subclasses can inherit diverse behaviors from various parent classes.

Modeling Complex Relationships: Useful for modeling real-world scenarios 
(e.g., a FlyingCar inheriting from both Car and Aircraft).

*Disadvantage:

Diamond Problem: When two base classes share a common ancestor, and a subclass inherits 
from both, ambiguity arises. Which method should the subclass use?

Method Resolution Order (MRO): Determining the order in which methods are called becomes 
crucial.

"""
#Example of multiple inheritance 

class Animal:
  def __init__(self, name):
    self.name = name

  def make_sound(self):
    print("Generic animal sound")

class Swimmer:
  def __init__(self):
    self.can_swim = True

  def swim(self):
    print("Swims in water")

class Dolphin(Animal, Swimmer):  # Inherits from both Animal and Swimmer
  def __init__(self, name):
    super().__init__(name)  # Call the first parent class constructor (Animal)

  def make_sound(self):
    print("Click, click!")  # Specific sound for dolphins

# Create a Dolphin object
dolphin = Dolphin("Flipper")
dolphin.make_sound()  # Output: Click, click! (from Dolphin)
dolphin.swim()        # Output: Swims in water (from Swimmer)
print(dolphin.name)   # Output: Flipper (from Animal)
```

```python
# Question 5. What is the purpose of the "super" keyword in inheritance? Provide an example.

"""
 "super" keyword two main purposes given below :-

1.Accessing Parent Class Features: 
It allows a subclass to refer to the parent class (also called the superclass) and access its attributes and 
methods. 
This is particularly useful when there's a naming conflict between the subclass and the parent class, 
or when you want to call the parent class's version of a method that's been overridden in the subclass.

2.Calling Parent Class Constructor:  
Within a subclass constructor, super() is used to explicitly call the constructor of the parent class. 
This ensures proper initialization of inherited attributes from the parent class.

"""
#Example

class Vehicle:
  def __init__(self, make, model):
    self.make = make
    self.model = model

  def info(self):
    print(f"This is a {self.make} {self.model}.")

class Car(Vehicle):
  def __init__(self, make, model, num_wheels):
    super().__init__(make, model)  # Call the parent class constructor
    self.num_wheels = num_wheels

  def info(self):  # Override the parent class info method
    print(f"This is a {self.make} {self.model} car with {self.num_wheels} wheels.")

# Create a Car object
car = Car("Toyota", "Camry", 4)
car.info()  # Output: This is a Toyota Camry car with 4 wheels.

```

```python
"""
Question 6. Create a base class called "Vehicle" with attributes like "make", "model", and "year".
Then, create a derived class called "Car" that inherits from "Vehicle" and adds an
attribute called "fuel_type". Implement appropriate methods in both classes.

"""

class Vehicle:
  """
  This class represents a generic vehicle.
  """
  def __init__(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year

  def info(self):
    """
    Prints basic information about the vehicle.
    """
    print(f"This is a {self.year} {self.make} {self.model}.")

class Car(Vehicle):
  """
  This class represents a car, inheriting attributes from Vehicle.
  """
  def __init__(self, make, model, year, fuel_type):
    super().__init__(make, model, year)  # Call parent class constructor
    self.fuel_type = fuel_type

  def info(self):
    """
    Prints detailed information about the car, overriding the parent class info method.
    """
    print(f"This is a {self.year} {self.make} {self.model} car with {self.fuel_type} fuel type.")

# Create a Vehicle object (not very useful on its own)
# vehicle = Vehicle("Generic", "Vehicle", 2024)

# Create a Car object
car = Car("Toyota", "Camry", 2023, "Hybrid")
car.info()  # Output: This is a 2023 Toyota Camry car with Hybrid fuel type.
```

```python
"""
Question 7. Create a base class called "Employee" with attributes like "name" and "salary."
Derive two classes, "Manager" and "Developer," from "Employee." Add an additional
attribute called "department" for the "Manager" class and "programming_language"
for the "Developer" class.

"""
class Employee:
  """
  This class represents a generic employee.
  """
  def __init__(self, name, salary):
    self.name = name
    self.salary = salary

class Manager(Employee):
  """
  This class represents a manager, inheriting attributes from Employee.
  """
  def __init__(self, name, salary, department):
    super().__init__(name, salary)  # Call parent class constructor
    self.department = department

class Developer(Employee):
  """
  This class represents a developer, inheriting attributes from Employee.
  """
  def __init__(self, name, salary, programming_language):
    super().__init__(name, salary)  # Call parent class constructor
    self.programming_language = programming_language

# Create Employee object (not very useful on its own)
# employee = Employee("John Doe", 50000)

# Create Manager object
manager = Manager("Jane Smith", 75000, "Engineering")

# Create Developer object
developer = Developer("Bob Jones", 60000, "Python")

print(f"Employee Name: {manager.name}")  # Output: Employee Name: Jane Smith (inherited from Employee)
print(f"Manager Department: {manager.department}")  # Output: Manager Department: Engineering
print(f"Developer Language: {developer.programming_language}")  # Output: Developer Language: Python
```

```python
"""
Question 8. Design a base class called "Shape" with attributes like "colour" and "border_width."
Create derived classes, "Rectangle" and "Circle," that inherit from "Shape" and add
specific attributes like "length" and "width" for the "Rectangle" class and "radius" for
the "Circle" class.

"""
class Shape:
  """
  This class represents a generic shape.
  """
  def __init__(self, colour, border_width):
    self.colour = colour
    self.border_width = border_width

class Rectangle(Shape):
  """
  This class represents a rectangle, inheriting attributes from Shape.
  """
  def __init__(self, colour, border_width, length, width):
    super().__init__(colour, border_width)  # Call parent class constructor
    self.length = length
    self.width = width

  def area(self):
    """
    Calculates and returns the area of the rectangle.
    """
    return self.length * self.width

class Circle(Shape):
  """
  This class represents a circle, inheriting attributes from Shape.
  """
  def __init__(self, colour, border_width, radius):
    super().__init__(colour, border_width)  # Call parent class constructor
    self.radius = radius

  def area(self):
    """
    Calculates and returns the area of the circle.
    """
    import math  # Import math library for pi
    return math.pi * self.radius * self.radius

# Create a Shape object (not very useful on its own)
# shape = Shape("red", 2)

# Create Rectangle object
rectangle = Rectangle("blue", 1, 5, 3)
print(f"Rectangle Area: {rectangle.area()}")  # Output: Rectangle Area: 15

# Create Circle object
circle = Circle("green", 1, 4)
print(f"Circle Area: {circle.area():.2f}")  # Output: Circle Area: 50.27 (round to 2 decimals)
```

```python
"""
Question 10. Create a base class called "BankAccount" with attributes like "account_number" and
"balance." Derive two classes, "SavingsAccount" and "CheckingAccount," from
"BankAccount." Add specific methods like "calculate_interest" for the
"SavingsAccount" class and "deduct_fees" for the "CheckingAccount" class.

"""
class BankAccount:
  """
  This class represents a generic bank account.
  """
  def __init__(self, account_number, balance):
    self.account_number = account_number
    self.balance = balance

  def deposit(self, amount):
    """
    Deposits a specified amount into the account balance.
    """
    self.balance += amount

  def withdraw(self, amount):
    """
    Withdraws a specified amount from the account balance, if sufficient funds exist.
    """
    if amount <= self.balance:
      self.balance -= amount
      return True
    else:
      return False

  def __str__(self):
    """
    Returns a string representation of the bank account.
    """
    return f"Account Number: {self.account_number}, Balance: ${self.balance:.2f}"

class SavingsAccount(BankAccount):
  """
  This class represents a savings account, inheriting attributes from BankAccount.
  """
  def __init__(self, account_number, balance, interest_rate):
    super().__init__(account_number, balance)  # Call parent class constructor
    self.interest_rate = interest_rate

  def calculate_interest(self):
    """
    Calculates and adds the interest earned on the balance to the account.
    """
    interest = self.balance * self.interest_rate
    self.deposit(interest)

class CheckingAccount(BankAccount):
  """
  This class represents a checking account, inheriting attributes from BankAccount.
  """
  def __init__(self, account_number, balance, monthly_fee):
    super().__init__(account_number, balance)  # Call parent class constructor
    self.monthly_fee = monthly_fee

  def deduct_fees(self):
    """
    Deducts the monthly fee from the account balance, if sufficient funds exist.
    """
    if self.monthly_fee <= self.balance:
      self.balance -= self.monthly_fee
      return True
    else:
      return False

# Create a BankAccount object (not very useful on its own)
# bank_account = BankAccount("12345678", 1000)

# Create SavingsAccount object
savings = SavingsAccount("87654321", 5000, 0.02)
savings.calculate_interest()
print(savings)  # Output: Account Number: 87654321, Balance: $5100.00

# Create CheckingAccount object
checking = CheckingAccount("98765432", 2000, 5)
if checking.deduct_fees():
  print(checking)  # Output: Account Number: 98765432, Balance: $1995.00
else:
  print("Insufficient funds for monthly fee.")
```
